## Outliers detection and removal using percentile in python pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('height.csv')
df.head()
```

```python
df.shape
```

### Detect outliers using percentile

```python
max_threshold = df['height'].quantile(0.95)
max_threshold
```

```python
df[df['height']> max_threshold]
```

```python
min_threshold = df['height'].quantile(0.05)
min_threshold
```

```python
df[df['height']< min_threshold]
```

### Remove outliers

```python
df1 = df[(df['height'] < max_threshold) & (df['height'] > min_threshold)]
df1
```

```python
df1.shape
```

### Using complex dataset

```python
df = pd.read_csv('AB_NYC_2019.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df['price'].describe()
```

```python
min_threshold, max_threshold = df.price.quantile([0.1, 0.999])
min_threshold, max_threshold
```

```python
df[df['price'] > max_threshold] 
```

```python
df[df.price < min_threshold]
```

```python
# remove ouliers
df_new = df[(df.price < max_threshold) & (df.price > min_threshold)]
df_new.sample(10)
```
